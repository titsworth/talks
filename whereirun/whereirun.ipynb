#Load the data

A quick search of google shows that the gpxpy package should do the trick to crunch the data. 

Let's build up a list of all of our data points from all GPX files. We index each run with a unique run index.

```python
def load_run_data(gpx_path):

    gpx_files = glob.glob(os.path.join(gpx_path,"*.gpx"))
    coords = []
    
    for idx, f in enumerate(gpx_files):     
        try :
            gpx_file = open(f, 'r')
            gpx = gpxpy.parse(gpx_file)
            
            track_coords = [[idx] + [point.latitude,point.longitude, point.elevation] 
                                for track in gpx.tracks 
                                    for segment in track.segments 
                                        for point in segment.points]
            
            coords += track_coords
        except:
            os.remove(f)
            
    return coords

coords = load_run_data(".")
```

# Plotting the data

I'll use pandas to group the data and plot each run as a seperate series.

```python
def make_segments(x, y):
    '''
    Create list of line segments from x and y coordinates, in the correct format for LineCollection:
    an array of the form   numlines x (points per line) x 2 (x and y) array
    '''

    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    
    return segments

def clear_frame(ax=None): 
    # Taken from a post by Tony S Yu
    if ax is None: 
        ax = plt.gca() 
    ax.xaxis.set_visible(False) 
    ax.yaxis.set_visible(False) 
    for spine in ax.spines.values(): 
        spine.set_visible(False)


def plot_run_data(coords, lat_bounds=(0,90), lon_bounds=(-180,0), show_elevation=False, cmap="Reds"):
    coords_df = pd.DataFrame(coords, columns=['Index','Latitude','Longitude','Altitude'])
    
    coords_df = coords_df[(coords_df['Latitude'] >= lat_bounds[0]) & (coords_df['Latitude'] <= lat_bounds[1])]
    coords_df = coords_df[(coords_df['Longitude'] >= lon_bounds[0]) & (coords_df['Longitude'] <= lon_bounds[1])]
                                                                      
    coords_df['Normalized_Elevation']=(coords_df['Altitude'] - coords_df['Altitude'].min())/(coords_df['Altitude'].max()-coords_df['Altitude'].min())
    
    grouped = coords_df.groupby('Index')
    
    fig = plt.figure(figsize=(15,13))
    fig.hold(True)
    
    ax = plt.gca()
    
    if show_elevation:
        mycmap = plt.get_cmap(cmap)
        segments = None
        elevations = None
        for name, group in grouped:
            new_segments = make_segments(group['Longitude'].values,group['Latitude'].values)
            if segments == None:
                segments = new_segments
                elevations = group['Normalized_Elevation'].values
            else:
                segments = np.concatenate([segments,new_segments])
                elevations = np.concatenate([elevations, group['Normalized_Elevation'].values])
        lc = LineCollection(segments, array=1-elevations, cmap=mycmap, norm=plt.Normalize(0.0, 1.0), linewidth=.5, alpha=.5)

        ax.add_collection(lc)
        
    else:
        grouped.plot('Longitude','Latitude', color='w', linewidth=.6, alpha=.5)

    plt.grid(False)
    plt.ylim(lat_bounds)
    plt.xlim(lon_bounds)
    plt.gca().patch.set_facecolor('#3B0062')
    
    clear_frame()
```
